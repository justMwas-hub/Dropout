```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dropout_dataset_final_with_target.csv")
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df
```

```python
X = df.drop("Target", axis=1)
y = df["Target"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
help(LogisticRegression)
```

```python
log_modelo = LogisticRegression(max_iter=10000)
```

```python
param_grid = {"penalty":["l1","l2","elasticnet"],
             "C":[0.001,0.005,0.01,0.02,0.05,0.1,0.2,0.5,0.99,1],
             "solver":['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
             "multi_class":['ovr', 'multinomial'],
             "l1_ratio":[0.001,0.01,0.1,0.5,0.99,1]}
```

```python
grid_modelo = GridSearchCV(log_modelo, param_grid=param_grid)
```

```python
grid_modelo.fit(X_train, y_train)
```

```python
grid_modelo.best_params_
```

```python
grid_modelo.best_score_
```

```python
grid_modelo.best_estimator_
```

```python
from sklearn.metrics import accuracy_score, plot_confusion_matrix, classification_report
```

```python
y_pred = grid_modelo.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
plot_confusion_matrix(grid_modelo,X_test,y_test)
```

```python
def run_model(model,X_train,y_train,X_test,y_test):
    
    # Fit Model
    model.fit(X_train,y_train)
    
    # Get Metrics
    
    preds = model.predict(X_test)
    
    print(classification_report(y_test,preds))
    print('\n')
    plot_confusion_matrix(model, X_test, y_test)
    
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
run_model(model,X_train,y_train,X_test,y_test)
```
